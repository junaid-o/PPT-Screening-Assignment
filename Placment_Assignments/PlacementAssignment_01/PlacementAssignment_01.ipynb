## Question 1:
**Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.**


```python
def find_highest_frequency_word_length(string):
    word_frequency = {}
    words = string.split()

    # Count the frequency of each word
    for word in words:
        # Remove repeated characters in the word
        #word = "".join(set(word))
        
        if word in word_frequency:
            word_frequency[word] += 1
        else:
            word_frequency[word] = 1

    # Find the word with the highest frequency
    highest_frequency = max(word_frequency.values())
    highest_frequency_word = [word for word, freq in word_frequency.items() if freq == highest_frequency]

    # Return the length of the highest-frequency word
    return highest_frequency, highest_frequency_word, len(highest_frequency_word[0])

                        
################################ TEST THE FUNCTION #################################

# Test the function
input_string = input("Enter a string: ")
highest_frequency, highest_frequency_word, highest_frequency_word_len = find_highest_frequency_word_length(input_string)

print("Highest-frequency:", highest_frequency)
print("Highest-frequency word:", highest_frequency_word)
print("Length of the highest-frequency word:", highest_frequency_word_len)
```

## Question 2:
**Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO**

```python
def is_valid_string(s):
    # Count the occurrence of each character in the string
    char_counts = {}
    for char in s:
        char_counts[char] = char_counts.get(char, 0) + 1

    # Count the occurrence of each count value
    count_counts = {}
    for count in char_counts.values():
        count_counts[count] = count_counts.get(count, 0) + 1

    # If there is only one count value, the string is already valid
    if len(count_counts) == 1:
        return "YES"

    # If there are more than two count values, the string can't be valid
    if len(count_counts) > 2:
        return "NO"

    # If there are exactly two count values, check if removing one character makes the string valid
    counts = list(count_counts.keys())
    freqs = list(count_counts.values())

    # Check if there is a single character that occurs only once
    if (freqs[0] == 1 and counts[0] == 1) or (freqs[1] == 1 and counts[1] == 1):
        return "YES"

    # Check if the difference between the two count values is exactly 1
    if abs(counts[0] - counts[1]) == 1 and (freqs[0] == 1 or freqs[1] == 1):
        return "YES"

    return "NO"

########################### TEST FUNCTION ##########################

s = input("Enter a string: ")
result = is_valid_string(s)
print(result)
```

### Question 10


This program uses the Natural Language Toolkit (NLTK) library to perform Part-of-Speech (POS) tagging on the input text. It then counts the occurrences of verbs (V), nouns (N), pronouns (P), and adjectives (J) using a Counter object. The result is returned as a dictionary where the keys are the POS tags and the values are their respective counts.

```python
import nltk
from nltk import pos_tag
from collections import Counter

def count_pos_tags(text):
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    # Perform Part-of-Speech (POS) tagging
    tagged_words = pos_tag(words)
    
    # Count the number of verbs, nouns, pronouns, and adjectives
    pos_counts = Counter(tag for word, tag in tagged_words if tag.startswith('V') or tag.startswith('N') or tag.startswith('P') or tag.startswith('J'))
    
    return pos_counts

# Example usage
phrase = "The quick brown fox jumps over the lazy dog."
counts = count_pos_tags(phrase)
print(counts)
```

```python
phrase = "She sells seashells by the seashore."
# Expected output: {'PRP': 1, 'VBZ': 1, 'NNS': 2}

counts = count_pos_tags(phrase)
print(counts)
```

```python
phrase = "I am happy and excited about the upcoming trip."
# Expected output: {'PRP': 1, 'VBP': 1, 'JJ': 2}

counts = count_pos_tags(phrase)
print(counts)
```

```python
phrase = "The cat chased the mouse."
# Expected output: {'DT': 2, 'NN': 2}

counts = count_pos_tags(phrase)
print(counts)
```

```python
phrase = "He is a diligent student who loves learning."
# Expected output: {'PRP': 1, 'VBZ': 1, 'DT': 1, 'JJ': 2, 'NN': 2}


counts = count_pos_tags(phrase)
print(counts)
```

###  Percentage of individuals with heights between 160 cm and 180 cm:

```python
import scipy.stats as stats

mean = 170
std_dev = 10

percentage = stats.norm.cdf(180, mean, std_dev) - stats.norm.cdf(160, mean, std_dev)
percentage *= 100
print(f"The percentage of individuals with heights between 160 cm and 180 cm is: {percentage}%")
```

### Probability that the average height of 100 individuals is greater than 175 cm:

```python
sample_size = 100
sample_mean = 175
sample_std_dev = std_dev / (sample_size ** 0.5)

probability = 1 - stats.norm.cdf(sample_mean, mean, sample_std_dev)
print(f"The probability that the average height of 100 individuals is greater than 175 cm is: {probability}")
```

### Z-score corresponding to a height of 185 cm:

```python
height = 185
z_score = (height - mean) / std_dev
print(f"The Z-score corresponding to a height of 185 cm is: {z_score}")
```

### Approximate height corresponding to the threshold where 5% of the dataset has heights below:

```python
threshold = stats.norm.ppf(0.05, mean, std_dev)
print(f"The approximate height corresponding to the threshold where 5% of the dataset has heights below is: {threshold}")
```

### Coefficient of Variation (CV) for the dataset:

```python
cv = (std_dev / mean) * 100
print(f"The coefficient of variation (CV) for the dataset is: {cv}%")
```
